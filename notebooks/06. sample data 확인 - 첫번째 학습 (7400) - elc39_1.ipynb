```python
from ast import literal_eval
from glob import glob
```

```python
data_dir = "./common/samples/elc39_1"

lines = []
files = [f for f in glob(data_dir + "/*", recursive=True)]
files.sort()
for f in files:
    print('getting [', f, "]...")
    with open(f, "r") as f:
        line = f.readline()
        while line:
            text = ''
            if line.startswith("="):
                line = f.readline()
                continue
            else:
                text += line
                line = f.readline()
            lines.append(text)
```

```python
samples = []
for i, sample in enumerate(lines):
    if sample != "\n":
        sample = sample.rstrip()
        sample = literal_eval(sample)
        samples.append(sample)
```
